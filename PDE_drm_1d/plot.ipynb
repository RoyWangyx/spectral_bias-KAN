```python
import numpy as np
from matplotlib import pyplot as plt
scales = [2,4,8,16,32]
losses = ['l2','h1']
model_choice = ['KAN','MLP']

for SCALE in scales:
    plt.figure(figsize=(7, 6))
    for loss in losses: 
        for model in model_choice:
            array = np.loadtxt(f'{loss}_scale_{SCALE}model_{model}.txt') 
            plt.plot(np.sqrt(array),label=f'{model}{loss}')
        
    plt.yscale('log')
    plt.xlabel('Training Iteration')
    plt.ylabel('Loss')
    plt.title(f"scale_{SCALE}")
    plt.legend()
    plt.savefig(f'scale_{SCALE}.png')
    plt.show()
```
