```python
import numpy as np
from matplotlib import pyplot as plt
depths = [2,3,4]
#depths = [6]
widths = [10]
#widths = [256]
grids = [10,20,30,40,50]
dims = [2,3,4]
scales = [0.125,0.25,0.5,1.0]
plt.rcParams.update({'axes.labelsize': 14, 'axes.titlesize': 16})
for DIM in dims:
    for SCALE in scales:
        model_choice = 'MLP'
        array = np.loadtxt(f'trainloss_dim_{DIM}_scale_{SCALE}_model_{model_choice}.txt')
        plt.figure(figsize=(7, 6))
        plt.plot(array,label='MLP_Tanh')
        model_choice = 'MLP_ReLU3'
        array = np.loadtxt(f'trainloss_dim_{DIM}_scale_{SCALE}_model_{model_choice}.txt')
        plt.plot(array,label=model_choice)
        #model_choice = 'MLP_ReLU1'
        #array = np.loadtxt(f'trainloss_dim_{DIM}_scale_{SCALE}_model_{model_choice}.txt')
        #plt.plot(array,label=model_choice)
        for DEPTH in depths:
            for WIDTH in widths:
                array = np.loadtxt(f'trainloss_depth{DEPTH}_dim_{DIM}_scale_{SCALE}_model_KAN.txt')
                plt.plot(array,label=f"KAN Width{WIDTH} Depth{DEPTH}")
        plt.yscale('log')
        plt.xlabel('Training Iteration')
        plt.ylabel('Loss')
        plt.title(f"Dim{DIM} Scale{SCALE}")
        plt.legend()
        plt.savefig(f'revision_trainDim{DIM}Scale{SCALE}.png')
        plt.show()
```
